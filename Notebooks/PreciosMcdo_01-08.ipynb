```python
import pandas as pd 
import matplotlib.pyplot as plt
import pylab
import numpy as np 
```

```python
def read_csv(name_dataset):
    df = pd.read_csv(name_dataset, sep=",", decimal=".")
    return df
    
```

```python
list_month = ['ENE','FEB','MAR','ABR','MAY','JUN','JUL','AGO']
def get_variacion(data, meses):
    data['variacion']= 0
    for i in range(len(meses)-1):
        #print data['ENE']
        a = np.abs(data[meses[i]] - data[meses[i+1]])
        data['variacion'] = a + data['variacion'] 
    return data
    
```

```python
df = read_csv('Workbook1.csv')
df = get_variacion(df,list_month)
df
```

```python
var = df[['ENE', 'FEB','MAR','ABR','MAY','JUN','JUL','AGO']].var(axis=1)
df['var']= var

df.sort_values('var', ascending=False).head(20)
```

# Productos con mayor variación ene - ago

Segun lo calculado, se observa que los 5 productos que más han variado en el periodo comprendido entre el mes de enero y el mes de agosto han sido la pitahaya (fruto tropical similar a la tuna), el zapallo loche, el jolantau (arvejas chinas), pacchoy (col asiática) y la linaza. Como algunos de estos productos no son de consumo frecuente en los hogares peruanos, se seleccionó productos más representativos:
- Zapallo loche
- Limón cítrico cajón
- Orégano
- Ajo pelado
- Ají amarillo seco
- Perejil nacional

```python
def plot_data(cod_prod,data):
    print data.iloc[cod_prod,2:10]
    ax = data.iloc[cod_prod,2:10].plot()
    ax.set_title('Precios por producto 2017')
    ax.set_ylabel('Soles')
    ax.legend([data.iloc[cod_prod,1]])
    plt.grid()
    plt.show()
```

# Productos con menor variación ene - ago

En contraste, los productos que menos variaron durante enero y agosto del 2017 fueron:
- Plátano bellaco
- Papa Huamantanga
- Lenteja seca nacional 
- Yuca amarilla 
- Azucar rubia nacional 

```python
def plot_data_dif(cod_prod1,cod_prod2,data):
    print data.iloc[cod_prod1,2:10]
    print data.iloc[cod_prod2,2:10]
    ax = data.iloc[cod_prod1,2:10].plot(legend=True)
    ax = data.iloc[cod_prod2,2:10].plot(legend=True)
    ax.set_title('Precios por producto 2017')
    ax.set_ylabel('Soles')
    ax.legend([[data.iloc[cod_prod1,1]][0],[data.iloc[cod_prod2,1]][0]])
    plt.grid()
    plt.show()
```

```python
plot_data_dif(95,74,df)
```
